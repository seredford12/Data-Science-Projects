```python
#import libraries and packages for working with dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
```

```python
#read in csv churn file
df = pd.read_csv("C:/Users/sered/OneDrive/Desktop/WGU Masters Degree/D212/teleco_market_basket.csv")
df.head()
```

```python
df.shape
```

```python
#Drop all rows with NaN values
df.dropna(axis = 0, how = 'all', inplace=True)
df.head()
```

```python
df.shape
```

```python
#Create a list of lists from the data
rows = []
for i in range (0,7501):
    rows.append([str(df.values[i,j])
for j in range (0,20)])
```

```python
#Feed the list created through transaction encoder to transform the data into a more usable form
DE = TransactionEncoder()
array = DE.fit(rows).transform(rows)

#Convert the array back to a datafram
df = pd.DataFrame(array, columns=DE.columns_)
```

```python
#Drop any columns with empty values after transforming the data
clean_df = df.drop(['nan'], axis = 1)
clean_df.head()
```

```python
#Save the cleaned data to a csv file
clean_df.to_csv('C:/Users/sered/OneDrive/Desktop/WGU Masters Degree/D212/Assessment 3/Cleaned_Data.csv')
```

```python
#import all necessary libraries to perform market basket analysis
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Create a visualization to determine the top products that customers buy
#Determine that top 5 items
count = clean_df.loc[:,:].sum()
pop_item = count.sort_values(axis = 0, ascending = False).head(5)
pop_item = pop_item.to_frame()
pop_item = pop_item.reset_index()
pop_item = pop_item.rename(columns = {'index': 'Products',0: 'count'})

#Create bar chart to show top items with purchase counts
plt.rcParams['figure.figsize'] = (10,6)
ax = pop_item.plot.barh(x = 'Products', y = 'count')
plt.title('Most Popular Products')
plt.gca().invert_yaxis()
```

```python
#Create Apriori object
obj = apriori(clean_df, min_support = 0.05, use_colnames = True)
obj.head()
```

```python
#Create the rules table for the apriori object
table = association_rules(obj, metric = 'lift', min_threshold = 1)
table.head(25)
```

```python
#Sorting Rules by metric
#Sort table by the objects with the highest confidence. Top 5 objects.

top_rules = table[(table['lift'] > 1.3) & (table['confidence'] > 0.2)].sort_values(by=['lift'], ascending= False)
top_rules
```
