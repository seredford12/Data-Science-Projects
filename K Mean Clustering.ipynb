```python
#define variables from the dataset to be used for analysis
df1 = df[['Lng','Lat', 'Outage_sec_perweek']]
df1.head()
```

```python
#import matplotlib and seaborn for visualizations
from matplotlib import pyplot as plt
import seaborn as sns


#Check for Outliers
df1[['Lng','Lat', 'Outage_sec_perweek']].hist()
plt.show()
```

```python
#plot the data for further exploration
sns.scatterplot(data = df1, x = 'Lng', y = 'Lat', hue='Outage_sec_perweek')
```

```python
#Normalize the dataframe for preprocessing
#import standard scaler from sklearn
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
scaled_df = scale.fit_transform(df1[['Lng','Lat', 'Outage_sec_perweek']])
```

```python
scaled_df = pd.DataFrame(scaled_df, columns=['Lng','Lat', 'Outage_sec_perweek'])
scaled_df.head()
```

```python
scaled_df.describe().round(2)
```

```python
#Save the cleaned and wrangled data to a CSV file
scaled_df.to_csv('C:/Users/sered/OneDrive/Desktop/WGU Masters Degree/D212/Assessment 1/processed_data.csv')
```

```python
#Import library to perform kmeans analysis
from sklearn.cluster import KMeans


#Create and fit KMeans model
k_model = KMeans(n_clusters = 5, n_init = 'auto', random_state = 42)

k_model.fit(scaled_df)
```

```python
#Visualize the data we just fit
sns.scatterplot(data = scaled_df, x = 'Lng', y = 'Lat', hue = k_model.labels_)
```

```python
#Evaluate the model
evaluate = pd.Series(k_model.labels_).value_counts()
print(evaluate)
```

```python
#visualize centroids to better explore and understand data
centroid = pd.DataFrame(k_model.cluster_centers_,
                        columns = ['Lng','Lat', 'Outage_sec_perweek'])
centroid
```

```python
#Plot figure showing clusters and centroids
ax = sns.scatterplot(data = scaled_df,
                     x = 'Lng',
                     y = 'Lat',
                     hue = k_model.labels_,
                     palette = 'colorblind',
                     alpha = 0.9,
                     s = 150,
                     legend = True)

ax = sns.scatterplot(data = centroid,
                     x = 'Lng',
                     y = 'Lat',
                     hue = centroid.index,
                     palette = 'colorblind',
                     s = 900,
                     marker = 'D',
                     ec = 'Black',
                     legend = False)

for i in range(len(centroid)):
    plt.text(x = centroid.Lng[i],
             y = centroid.Lat[i],
             s = i,
             horizontalalignment = 'center',
             verticalalignment = 'center',
             size = 15,
             weight = 'bold',
             color = 'white')
```

```python
#Determine optimal number of clusters
wcss = []
for k in range (1,10):
    model = KMeans(n_clusters = k, n_init = 'auto', random_state = 42)
    model.fit(scaled_df)
    wcss.append(model.inertia_)
wcss_s = pd.Series(wcss, index = range(1,10))
    
plt.figure(figsize=(12,10))
ax = sns.lineplot(y = wcss_s, x = wcss_s.index)
ax = sns.scatterplot(y = wcss_s, x = wcss_s.index, s = 200)
ax = ax.set(xlabel = 'Optimal k clusters',
            ylabel = 'Within Clusters Sum of Squares (wcss)')
```

```python
#Determine silhousette score to evaluate consistency of clusters
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(scaled_df, k_model.labels_)
sil_score
```

```python
#perform sillouette analysis to determine optimal k value
sil = []

for k in range (2,10):
    model = KMeans(n_clusters = k, n_init = 'auto', random_state = 42)
    model.fit(scaled_df)
    sil.append(silhouette_score(scaled_df, model.labels_))

sil_s = pd.Series(sil, index = range(2,10))

plt.figure(figsize=(12,10))
ax = sns.lineplot(y = sil_s, x = sil_s.index)
ax = sns.scatterplot(y = sil_s, x = sil_s.index, s = 200)
ax = ax.set(xlabel = 'Optimal k Number',
            ylabel = 'Silhouette Score Avg.')
```

```python
#Create optimized model with the k value found
opt_model = KMeans(n_clusters = 4, n_init = 'auto', random_state = 42)

opt_model.fit(scaled_df)
```

```python
#Visualize the optimized model
sns.scatterplot(data = scaled_df, x = 'Lng', y = 'Lat', hue = opt_model.labels_)
```

```python
#Evaluate the model
evaluate = pd.Series(opt_model.labels_).value_counts()
print(evaluate)
```

```python
#visualize centroids to better explore and understand optimized model
centroid = pd.DataFrame(opt_model.cluster_centers_,
                        columns = ['Lng','Lat', 'Outage_sec_perweek'])
centroid
```

```python
#Plot figure showing clusters and centroids
ax = sns.scatterplot(data = scaled_df,
                     x = 'Lng',
                     y = 'Lat',
                     hue = opt_model.labels_,
                     palette = 'colorblind',
                     alpha = 0.9,
                     s = 150,
                     legend = True)

ax = sns.scatterplot(data = centroid,
                     x = 'Lng',
                     y = 'Lat',
                     hue = centroid.index,
                     palette = 'colorblind',
                     s = 900,
                     marker = 'D',
                     ec = 'Black',
                     legend = False)

for i in range(len(centroid)):
    plt.text(x = centroid.Lng[i],
             y = centroid.Lat[i],
             s = i,
             horizontalalignment = 'center',
             verticalalignment = 'center',
             size = 15,
             weight = 'bold',
             color = 'white')
```

```python
#Optimized model silhouette score to understand cluster consistency
sil_score_opt = silhouette_score(scaled_df, opt_model.labels_)
sil_score_opt
```
